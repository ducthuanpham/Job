```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x[0] - x[1]**2 - 3)**2 + (x[0] - x[1] - 1)**2

def gradient_f(x):
    df_dx1 = 2 * (x[0] - x[1]**2 - 3) + 2 * (x[0] - x[1] - 1)
    df_dx2 = -4 * x[1] * (x[0] - x[1]**2 - 3) - 2 * (x[0] - x[1] - 1)
    return np.array([df_dx1, df_dx2])

def gradient_descent(x0, lr, steps = 10):
    trajectory = [x0]
    x = x0
    for _ in range(steps):
        x = x - lr * gradient_f(x)
        trajectory.append(x)
    return np.array(trajectory)

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f([X, Y])

def contours_plot(lrate, clr):
    plt.figure(figsize=(10, 6))
    plt.contour(X, Y, Z, levels=50)
    plt.colorbar(label='f(x)')
    plt.title('Contour Plot of f(x) with  = '+str(lrate))
    plt.xlabel('x1')
    plt.ylabel('x2')
    tr= gradient_descent(np.array([-0.5, -0.5]), lr=lrate)
    plt.plot(tr[:, 0], tr[:, 1], marker='o', label=' =' +str(lrate), color=clr)
    plt.legend()
    plt.grid(True)
    plt.show()


contours_plot(0.01,'red')
contours_plot(0.05,'blue')
```
